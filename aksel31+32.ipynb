### 3.1

```python
c, m, ks, delta, f = symbols("c m ks delta f")
lign1 = Eq(-m*c+ks*(1+tan(delta)*c), f)
display(lign1)
solve(lign1, c)
```

### 3.2

```python
f, ks, delta, m, w, t = symbols('f ks delta m omega t')
```

```python
x_p = (f-ks)/(ks*tan(delta)-m)*cos(w*t)
```

```python
def plot_x_p(omega):
    prepared_x_p = x_p.subs({m:3, ks:2, f:1, delta:.3, w:omega})
    return dtuplot.plot(prepared_x_p, show=False, xlim=(-2, 2), label=f"Omega = {omega}")
```

```python
comb = sum([plot_x_p(o) for o in range(1, 6)])
comb.show()
```
